# Matrix exponent


To begin, we will practice finding the eigenvalues and eigenvectors of a matrix. For this, we will work with polynomials using the sympy library.

```python
A = np.array([[1, 2, 3],[0, 2, 3],[0, 0, 3]])
I = np.identity(A.shape[0])

# A, I
```

```python
x = sympy.Symbol('x')
A_sym = sympy.Matrix(A)
I_sym = sympy.Matrix(I)
matrix_expr = A_sym - x * I_sym

# A_sym, I_sym, matrix_expr
```

```python
char_poly = matrix_expr.det()
roots = sympy.solve(char_poly, x)

# char_poly, roots
```

```python
for root in roots:
    A_minus_lambda_I = A_sym - root * I_sym
    null_space_basis = A_minus_lambda_I.nullspace()
    print(f"Eigenvalue = {root}, \neigenvectors:")
    for vector in null_space_basis:
        print(vector)
```
